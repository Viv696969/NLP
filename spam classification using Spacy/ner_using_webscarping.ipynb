```python
from time import sleep
```

```python
def give_data(url=None):
    try:
        if url:
            data=requests.get(url)
            soup=bs4.BeautifulSoup(data.content,'html.parser')
            # print(soup.prettify)
            data=str(soup.find_all("script")[4])
            data=re.findall(pattern='"articleBody": "([\(\)?$a-zA-Z0-9., &;-]*) ",\n  "publisher":',string=data)[0]
            return data
        else:
            return None
    except:
        print("Unknown error identified")
        return None
```

```python
def display_entities(data):
    try:
        nlp=spacy.load("en_core_web_sm")
        doc=nlp(data)
        explaination_dict={}
        for entity in doc.ents:
            if entity.label_ not in explaination_dict:
                explaination_dict[entity.label_]=spacy.explain(entity.label_)
    
        print(explaination_dict)
        sleep(5)
        print()
        spacy.displacy.render(doc,style="ent")
    except:
        print("unknown error")
    
    
    
```

```python
data=give_data('https://www.zeebiz.com/markets/global-markets/news-asian-markets-news-stocks-surge-on-chinas-spending-plans-inflation-shocks-australia-261092')
if data !=None:
    display_entities(data)
```
